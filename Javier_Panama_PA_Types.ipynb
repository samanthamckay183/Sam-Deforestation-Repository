```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()
```

```python
Map = geemap.Map()

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")
panama_fc = countries.filter(ee.Filter.eq("ADM0_NAME", "Panama"))
panama_geom = panama_fc.geometry()

Map.centerObject(panama_geom, 7)
```

### SINIA Landuse and Forest Cover 2021

```python
sinia_prism = ee.Image("projects/deforestation-panama/assets/landuse_2021").clip(panama_geom)

visParams = {
    "min": 1,
    "max": 33, 
    "palette": [
        '32a65e',  # [1] Mature mixed broadleaf forest
        '32a65e',  # [2] Secondary mixed broadleaf forest
        '1f8d49',  # [3] Mangrove forest
        '7dc975',  # [4] Orey Forest
        '04381d',  # [5] Cativo Forest
        '026975',  # [6] Raffia forest
        '000000',  # [7] Planted coniferous forest
        '000000',  # [8] Planted forest of broadleaf trees
        '7a6c00',  # [9] Stubble and shrub vegetation
        'ad975a',  # [10] Herbaceous vegetation
        '519799',  # [11] Low-lying, floodable vegetation
        'd6bc74',  # [12] Rocky outcrop and bare earth
        'd89f5c',  # [13] Beach and natural sand dune
        'FFFFB2',  # [14] Coffee
        'edde8e',  # [15] Citrus
        '000000',  # [16] Oil palm
        '000000',  # [17] Plantain/banana
        'f5b3c8',  # [18] Another permanent crop
        'C27BA0',  # [19] Rice
        'db7093',  # [20] Sugar cane
        'ffefc3',  # [21] Mixed horticulture
        'db4d4f',  # [22] Corn
        'ffa07a',  # [23] Pineapple
        'd4271e',  # [24] Another annual crop
        'db4d4f',  # [25] Heterogeneous area of ​​agricultural production
        '0000FF',  # [26] Grass
        '000000',  # [27] Water surface
        '000000',  # [28] Populated area
        'ffaa5f',  # [29] Infrastructure
        '9c0027',  # [30] Mining
        '091077',  # [31] Aquaculture pond
        'fc8114',  # [32] Saltworks
        '2532e4',  # [33] Albino
    
    ]
}

Map = geemap.Map()
Map.centerObject(panama_geom, 7)

Map.addLayer(sinia_prism, visParams, "SINIA Landuse")

Map
```

### Prism ESA Land Cover Classification - Plant Functional Types

```python
esa_prism = ee.Image("projects/deforestation-panama/assets/esa_landuse").select(['b1']).clip(panama_geom)
```

```python
esa_prism.projection().nominalScale().getInfo()
```

```python
esa_prism.bandNames().getInfo()
```

```python
Map = geemap.Map()
Map.centerObject(panama_geom, 7)

Map.addLayer(esa_prism, {'min': 40, 'max': 65, 'palette': ['white', 'green']}, "ESA landcover class")

Map
```

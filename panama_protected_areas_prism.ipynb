```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()

# Create map
Map = geemap.Map(center=[8.5, -80], zoom=7)

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")

panama = countries.filter(
    ee.Filter.eq("ADM0_NAME", "Panama")
)

pa = ee.FeatureCollection("projects/deforestation-495419/assets/panama_protected_areas_polygons")

Map.addLayer(panama)
Map.addLayer(pa, {}, 'panama')

# Display map
Map
```

```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

# Create map
Map = geemap.Map(center=[8.5, -80], zoom=7)

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")

panama = countries.filter(
    ee.Filter.eq("ADM0_NAME", "Panama")
)

# Units in meters
basins = ee.Image("projects/deforestation-495419/assets/hydrographic_basins")

# Clip to Panama
# slope_clipped = slope.clip(panama)

Map.addLayer(panama)
Map.addLayer(basins, {}, 'hydrographic basins')

# Display map
Map
```

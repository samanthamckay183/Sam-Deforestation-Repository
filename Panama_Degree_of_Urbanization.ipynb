```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

# Load GHSL SMOD dataset
image = ee.Image("JRC/GHSL/P2023A/GHS_SMOD_V2-0/2030")
smod = image.select('smod_code')

# Get Panama boundary
panama = (
    ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017")
    .filter(ee.Filter.eq("country_na", "Panama"))
    .geometry()
)

# Clip image to Panama
image_panama = image.clip(panama)

# Create map
Map = geemap.Map()
Map.setCenter(-80, 8.5, 5)

# Add layer
Map.addLayer(image_panama, {}, "Degree of Urbanization (Panama)")

Map
```

```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

# Create map
Map = geemap.Map(center=[8.5, -80], zoom=7)

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")

panama = countries.filter(
    ee.Filter.eq("ADM0_NAME", "Panama")
)

# in dg/kg
pa = ee.Image("projects/deforestation-495419/assets/Soil_Org_C")
# in cg/kg
pn = ee.Image("projects/deforestation-495419/assets/Soil_N")
# in g/kg
ps = ee.Image("projects/deforestation-495419/assets/sand")

Map.addLayer(panama)
Map.addLayer(pa, {}, 'soil_organic_carbon')
Map.addLayer(pn, {}, 'nitrogen')
Map.addLayer(ps, {}, 'sand')

# Display map
Map
```

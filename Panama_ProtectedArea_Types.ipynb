```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()
```

```python
Map = geemap.Map()

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")
panama_fc = countries.filter(ee.Filter.eq("ADM0_NAME", "Panama"))
panama_geom = panama_fc.geometry()

Map.centerObject(panama_geom, 7)
```

```python
pa_polygons = ee.FeatureCollection("projects/deforestation-panama/assets/WDPA_WDOECM_Jul2026_Public_PAN_shp-polygons")

```

```python
Map.addLayer(pa_polygons, {}, "PA polygons")

Map
```

```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

# Load Panama boundary
panama = (
    ee.FeatureCollection('FAO/GAUL/2015/level0')
    .filter(ee.Filter.eq('ADM0_NAME', 'Panama'))
)

# Load ERA5 hourly dataset
dataset = (
    ee.ImageCollection('ECMWF/ERA5/HOURLY')
    .filter(ee.Filter.date('2020-07-01', '2020-07-02'))
)

# Clip each image to Panama
temperature = dataset.map(lambda img: img.clip(panama))

# Visualization parameters
visualization = {
    'bands': ['temperature_2m'],
    'min': 250.0,
    'max': 320.0,
    'palette': [
        '000080', '0000d9', '4000ff', '8000ff',
        '0080ff', '00ffff', '00ff80', '80ff00',
        'daff00', 'ffff00', 'fff500', 'ffda00',
        'ffb000', 'ffa400', 'ff4f00', 'ff2500',
        'ff0a00', 'ff00ff',
    ]
}

# Create map
m = geemap.Map()

# Center on Panama
m.centerObject(panama, 7)

# Add clipped ERA5 layer
m.add_layer(
    temperature,
    visualization,
    'Air temperature [K] at 2m height'
)

# Add Panama boundary
m.add_layer(panama, {}, 'Panama Boundary')

m
```

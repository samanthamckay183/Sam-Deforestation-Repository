```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

dataset = ee.Image('UMD/hansen/global_forest_change_2025_v1_13')

tree_cover_vis_param = {
    'bands': ['treecover2000'],
    'min': 0,
    'max': 100,
    'palette': ['black', 'green'],
}

tree_loss_vis_param = {
    'bands': ['lossyear'],
    'min': 0,
    'max': 25,
    'palette': ['yellow', 'red'],
}

m = geemap.Map(center=[8.5, -80], zoom=7)
m.add_layer(dataset, tree_cover_vis_param, 'tree cover')
m.add_layer(dataset, tree_loss_vis_param, 'tree loss year')
m
```

```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()

# Create map
Map = geemap.Map(center=[8.5, -80], zoom=7)

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")

panama = countries.filter(
    ee.Filter.eq("ADM0_NAME", "Panama")
)

roads = ee.FeatureCollection("projects/deforestation-495419/assets/Panama_OSM_Roads")

# Rasterize roads
roads_raster = ee.Image().byte().paint(roads, 1)

# Distance to nearest road in meters
distance_to_roads = roads_raster.distance(
    ee.Kernel.euclidean(5000, 'meters')
).clip(panama)

Map.addLayer(panama)
Map.addLayer(roads, {}, 'panama')
Map.addLayer(
    distance_to_roads,
    {
        'min': 0,
        'max': 5000,
        'palette': ['white', 'blue', 'green', 'yellow', 'red']
    },
    'Distance to Roads'
)

# Display map
Map
```

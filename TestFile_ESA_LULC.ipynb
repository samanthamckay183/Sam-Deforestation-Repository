```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()
```

```python
Map = geemap.Map()

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")
panama_fc = countries.filter(ee.Filter.eq("ADM0_NAME", "Panama"))
panama_geom = panama_fc.geometry()

Map.centerObject(panama_geom, 7)
```

### SINIA Landuse and Forest Cover

```python
sinia_prism = ee.Image("projects/deforestation-panama/assets/landuse_2021").clip(panama_geom)
```

```python
sinia_prism.projection().nominalScale().getInfo()
```

```python
sinia_prism.bandNames().getInfo()
```

```python
print(
    sinia_prism.reduceRegion(
        reducer=ee.Reducer.frequencyHistogram(),
        geometry=panama_geom,
        scale=30,
        maxPixels=1e13
    ).getInfo()
)
```

```python
Map = geemap.Map()
Map.centerObject(panama_geom, 7)

Map.addLayer(sinia_prism, {'min': 0, 'max': 33, 'palette': ['white', 'green']}, "SINIA Landuse")

Map
```

### Prism ESA Land Cover Classification

```python
esa_prism = ee.Image("projects/deforestation-panama/assets/esa_landuse").select(['b1']).clip(panama_geom)
```

```python
esa_prism.projection().nominalScale().getInfo()
```

```python
esa_prism.bandNames().getInfo()
```

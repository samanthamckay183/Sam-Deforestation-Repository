```python
import ee 
import geemap
ee.Authenticate()
ee.Initialize()

# Create interactive map centered on Panama
Map = geemap.Map(center=[8.5, -80], zoom=7)

# Load Panama boundary from FAO GAUL dataset
panama = (
    ee.FeatureCollection('FAO/GAUL/2015/level0')
    .filter(ee.Filter.eq('ADM0_NAME', 'Panama'))
)

# Load drivers dataset
drivers = ee.Image(
    'projects/landandcarbon/assets/wri_gdm_drivers_forest_loss_1km/v1_2_2001_2024'
)

# Clip image to Panama
drivers_panama = drivers.clip(panama)

# Select classification band
drivers_class = drivers_panama.select(['classification'])

# Visualization parameters
vis = {
    'min': 1,
    'max': 7,
    'palette': [
        'E39D29',
        'E58074',
        'e9d700',
        '51a44e',
        '895128',
        'a354a0',
        '3a209a'
    ]
}

# Add classification layer
Map.addLayer(
    drivers_class,
    vis,
    'Drivers of Forest Loss, Panama'
)

# Select probability band
permAg_prob = drivers_panama.select(['probability_1'])

# Visualization parameters for probability layer
probVis = {
    'min': 0,
    'max': 250,
    'palette': [
        '#440154',
        '#481567',
        '#482677',
        '#453781',
        '#3b528b',
        '#2c728e',
        '#21908d',
        '#27ad81',
        '#5ec962',
        '#aadc32',
        '#fde725'
    ]
}

# Add probability layer (hidden by default)
Map.addLayer(
    permAg_prob,
    probVis,
    'Probability of Permanent Agriculture',
    shown=False
)

# Add Panama boundary
Map.addLayer(panama, {}, 'Panama Boundary')

# Display map
Map
```

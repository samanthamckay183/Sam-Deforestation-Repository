## WAPOR Actual Evapotranspiration and Interception 3.0

```python
import ee
import geemap
ee.Authenticate()
ee.Initialize()
```

```python
Map = geemap.Map()

# Panama boundary
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")
panama_fc = countries.filter(ee.Filter.eq("ADM0_NAME", "Panama"))
panama_geom = panama_fc.geometry()

Map.centerObject(panama_geom, 7)
```

```python
# Actual Evapotranspiration and Interception (Dekadal) [mm]
coll = ee.ImageCollection('FAO/WAPOR/3/L1_AETI_D').first().clip(panama_geom)

Map = geemap.Map()
Map.centerObject(panama_geom, 7)
Map.addLayer(coll, {'min': 0, 'max': 50}, 'Evapotranspiration')

Map
```
